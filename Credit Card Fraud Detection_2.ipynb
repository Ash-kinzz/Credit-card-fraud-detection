```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
```

```python
# loading the dataset to a Pandas DataFrame
credit_card_data = pd.read_csv('creditcard.csv')
```

```python
# checking the number of missing values in each column
credit_card_data.isnull().sum()
```

```python
# Handle missing values (example using mean imputation)
imputer = SimpleImputer(strategy="mean")
X = imputer.fit_transform(credit_card_data.drop('Class', axis=1))
```

```python
# distribution of legit transactions & fraudulent transactions
credit_card_data['Class'].value_counts()
```

```python
# separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
# Feature scaling
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
# Separate target variable
Y = credit_card_data['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Train Random Forest model
model = RandomForestClassifier(n_estimators=100, random_state=42)  # Adjust hyperparameters as needed
model.fit(X_train, Y_train)
```

```python
# Make predictions on training and test data
X_train_prediction = model.predict(X_train)
X_test_prediction = model.predict(X_test)
```

```python
# Evaluate model performance
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
precision = precision_score(Y_test, X_test_prediction)
recall = recall_score(Y_test, X_test_prediction)
f1 = f1_score(Y_test, X_test_prediction)
roc_auc = roc_auc_score(Y_test, X_test_prediction)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
print('Accuracy score on Test Data : ', test_data_accuracy)
print('Precision : ', precision)
print('Recall : ', recall)
print('F1 Score : ', f1)
print('ROC AUC Score : ', roc_auc)
```
